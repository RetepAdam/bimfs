```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/cbb_2001_to_2017.csv')
df_nba = pd.read_csv('data/rookies3.csv')
```

```python
df_remaining = df[df['Player'].isin(df_nba['Player'])].sort_values('Player')
```

```python
df_try = df_remaining.groupby(['Player', 'School']).sum()
```

```python
df_try.reset_index(inplace=True)
```

```python
df_try.to_csv('data/try.csv')
```
